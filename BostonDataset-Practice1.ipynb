```python
# Training the  model
```

```python
# Independent and Dependent Features
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
# Independent Feature
X.head()
```

```python
# Dependent Feature
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.33, random_state=10)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
# Standardize or feature scaling the dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
# Why transform - to prevent from data leakage
```

```python
X_train
```

```python
X_test
```

```python
# search for data leakage
```

```python
# Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
# We are applying formula not changing data
regression.fit(X_train,Y_train)
```

```python
# Linear Regression
```

```python
# Print the coefficient and the intercept|
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# if all the independet feature values is 0,the price is 22.07
```

```python
# Prediction
```

```python
reg_pred = regression.predict(X_test)
reg_pred
```

```python
# Assumptions
# The relationship between actual and predicted data should be linear
```

```python
plt.scatter(Y_test,reg_pred)
plt.xlabel("Test Actual Data")
plt.ylabel("Test Predicted Data")
```

```python
residuals = Y_test - reg_pred
residuals
```

```python
sns.distplot(residuals,kde=True)
```

```python
# The distribution should be uniform (Homoscedasticity)-does not have any kind of shape
plt.scatter(reg_pred,residuals)
```

```python
# Performance Matrix
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(Y_test,reg_pred))
print(mean_absolute_error(Y_test,reg_pred))
print(np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

```python
# R-Squared and Adjusted R-Squared
from sklearn.metrics import r2_score
score = r2_score(Y_test,reg_pred)
print(score)
```

```python
# Adjusted R square
1-(1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```
